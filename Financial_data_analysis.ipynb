```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Net Present Value (NPV) Analysis for Portfolio

# Update Project Greenspace Cash Flows Here
project_a = [-2000000, 0, 0, 50000, 250000, 500000, 500000, 750000, 750000, 1000000]

discount_rate = [0.05, 0.055, 0.06, 0.065, 0.07, 0.075, 0.08, 0.085, 0.09, 0.095, 0.1, 0.105, 0.11, 0.115, 0.12, 0.125, 0.13, 0.135, 0.14, 0.145, 0.15]

def calculate_npv(rate, cash_flow):
    npv = 0
    for t in range(len(cash_flow)):
        npv += cash_flow[t]/(1+rate)**t
    return npv

npvs_a = list()
for rate in discount_rate:
  npv_a = calculate_npv(rate,project_a)
  npvs_a.append(npv_a)
  
plt.plot(discount_rate,npvs_a, linewidth = 3.0, color = "green", label = "Project Greenspace")
plt.axhline(y=0, linewidth = 0.5, color = "black")
plt.title('NPV Profile: Project Greenspace ')
plt.xlabel('Discount Rate')
plt.ylabel('Net Present Value')
plt.legend()
plt.show()
```

# Simple Rate of Return

```python
# defining the function to convert to percentage

def display_as_percentage(value):
    return '{:.1f}%'.format(value * 100)
```

```python
# defining the function to Calculate the Simple Return on investment

def calculate_simple_return(start_price, end_price, dividend=0):
    return (end_price - start_price + dividend) / start_price
```

```python
return_on_investment = calculate_simple_return(100000, 102000, 5)

print(f'The rate of return on investment is : {display_as_percentage(return_on_investment)}')
```

# Logarithmic Rate of Return

```python
# Import library here
from math import log

def display_as_percentage(val):
    return '{:.1f}%'.format(val * 100)

# Write code here

def calculate_log_return(start_price, end_price):
    return log(end_price/start_price)

log_return = calculate_log_return(200, 250)

print(f'The log rate of return is {display_as_percentage(log_return)}')

```

# Aggregate Across Time 


#### If we are converting daily returns to annual returns, 't' may be 252 because that is typically the number of trading days in a year. If we are converting monthly returns to annual returns, 't' would be 12

```python
def display_as_percentage(val):
    return '{:.1f}%'.format(val * 100)

daily_return_a = 0.001
monthly_return_b = 0.022

```

```python
print(f'The daily rate of return for Investment A is {display_as_percentage(daily_return_a)}')
print(f'The monthly rate of return for Investment A is {display_as_percentage(monthly_return_b)}')
```

```python
def annualize_return(log_return, t):
    return log_return * t
```

```python
annual_return_a = annualize_return(daily_return_a, 252)
print(f'The annual rate of return for investment A is {display_as_percentage(annual_return_a)}')
```

```python
annual_return_b = annualize_return(monthly_return_b, 12)
print(f'The annual rate of return for Investment B is {display_as_percentage(annual_return_b)}')
```

```python
def display_as_percentage(val):
    return '{:.1f}%'.format(val * 100)

daily_returns = [0.002, -0.002, 0.003, 0.002, -0.001]

# defining the function to convert returns by using the formula from above

def convert_returns(log_returns, t):
    return (sum(log_returns) / len(log_returns) ) * t
```

```python
annual_return = convert_returns(daily_returns, 252)


print('The annual rate of return is', display_as_percentage(annual_return))
```

```python
weekly_return = convert_returns(daily_returns, 5)
print('The weekly rate of return is', display_as_percentage(weekly_return))
```

## Variance
- ***Variance is a measure of the spread of a dataset, or how far apart each value is from the mean.***
- ***The greater the variance, the more spread out or variable the data is***

```python
import numpy as np

# Suppose we have investment returns on these two over the course of a week
# the Investment returns on disney fluctuates more than cbs. Therefore it is considered more volatile and riskier. 
returns_disney = [0.22, 0.12, 0.01, 0.05, 0.04]
returns_cbs = [-0.13, -0.15, 0.31, -0.06, -0.29]

variance_disney = np.var(returns_disney)
variance_cbs = np.var(returns_cbs)

```

```python
dataset = [10, 8, 9 ,10, 12]

mean = np.mean(dataset)
print(mean)
```

```python
# Calculating the sum in the nominator of the variance formula by looping through the 'dataset' list and
# adding up the sqaured difference between each data point and the 'mean'

numerator = 0

for i in dataset:
    numerator += np.power(i - mean, 2)

print(numerator)
```

```python
# Divide the 'numerator' by the number of values in 'dataset' (len)

variance = numerator / len(dataset)

print(variance)
```

```python
# Now, combine calculation from above into a function

def calculate_variance(dataset):
    mean = np.mean(dataset)
    
    numerator = 0
    for i in dataset:
        numerator += np.power(i - mean, 2)
    
    variance = numerator / len(dataset)
    
    return variance
```

```python
# Using the function above to reassign the values to "variance_disney and *_cbs"

variance_disney = calculate_variance(returns_disney)
variance_cbs = calculate_variance(returns_cbs)

print(f'The variance of Disney stock returns is {variance_disney}')
print(f'The variance of CBS stock returns is {variance_cbs}')

# after using the formula , 
# this concludes that cbs stock has greater variance and its returns are less consistent and riskier
# The conclusing contradicts my opinion when i looked at the returns without using the variance formula.
```

# Standard Deviation 

***Although the variance is useful in determing the relative risk of an investment, it is sometimes not the easiest statistic to interpret since it does not have the same unit as the original data. This is where Std comes into play to describe the spread of the dataset***

***Standard Deviation is simply the square root of the variance***

```python
def display_as_percentage(val):
    return '{:.1f}%'.format(val * 100)
```

```python
stddev_disney = np.std(returns_disney)
stddev_cbs = np.std(returns_cbs)

print('The standard deviation of Disney stock returns is', display_as_percentage(stddev_disney))
print('The standard deviation of CBS stock returns is', display_as_percentage(stddev_cbs))
```
